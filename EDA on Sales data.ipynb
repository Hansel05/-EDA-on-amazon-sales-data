```python
def rating_func(row):
    if row['rating'] >= 2.0 and row['rating'] <= 2.4:
        return 'Unsatisfactory'
    elif row['rating'] >= 2.5 and row['rating'] <= 2.9:
        return 'Needs Improvement'
    elif row['rating'] >= 3.0 and row['rating'] <= 3.4:
        return 'Satisfactory'
    elif row['rating'] >= 3.5 and row['rating'] <= 3.9:
        return 'Average'
    elif row['rating'] >= 4.0 and row['rating'] <= 4.4:
        return 'Good'
    else :
        return 'Excellent'

df['rating_cat'] = df.apply(rating_func, axis = 1)
```

```python
rating_array = np.array(['2.0 - 2.4', '2.5 - 2.9', '3.0 - 3.4', '3.5 - 3.9', '4.0 - 4.4', '4.5 - 5.0'])
rating_desc = np.array(['Unsatisfactory', 'Needs Improvement', 'Satisfactory', 'Average', 'Good', 'Excellent'])
rating_map = pd.DataFrame(data = [rating_array, rating_desc]).T
rating_map.columns = ['rating', 'description']
rating_map
```
